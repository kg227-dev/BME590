```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Accuracies in the order (Random Forest, SVM, XGBoost, Ensemble)
acc_random = [0.9520958083832335, 0.9640718562874252, 0.9560878243512974,0.9600798403193613]
acc_scaffold = [0.8413173652694611,0.8922155688622755, 0.8802395209580839,0.8832335329341318]
```

```python
#Balanced Accuracies in the order (Random Forest, SVM, XGBoost, Ensemble)
bal_acc_random = [0.88,0.9212593516209476,0.9162718204488778,0.907506234413965]
bal_acc_scaffold = [0.8172413793103448,0.8766648421820835,0.8628717387338076,0.8655172413793104]
```

```python
#ROC AUC in the order (Random Forest, SVM, XGBoost, Ensemble)
roc_auc_random = [0.88, 0.9212593516209476, 0.9162718204488777, 0.9075062344139652 ]
roc_auc_scaffold = [0.8172413793103448,0.8766648421820835, 0.8628717387338077,0.8655172413793104]
```

```python
#Precision in the order (Random Forest, SVM, XGBoost, Ensemble)
precision_random = [0.9435294117647058,0.9636803874092009, 0.9633251833740831,0.9568345323741008]
precision_scaffold = [0.78099173553719,0.8430493273542601, 0.8281938325991189, 0.8289473684210527]
```

```python
#Recall in the order (Random Forest, SVM, XGBoost, Ensemble)
recall_random = [1.0, 0.9925187032418953, 0.9825436408977556, 0.9950124688279302]
recall_scaffold = [1.0, 0.9947089947089947, 0.9947089947089947, 1.0]
```

```python
# Please write your code below:
metrics = ['Accuracy', 'Balanced accuracy','ROC_AUC', 'Precision', 'Recall']
plt.figure(figsize=(20,10))
plt.subplot(polar=True)
theta = np.linspace(0,2*np.pi, 4)
lines, labels = plt.thetagrids(range(0, 360, int(360/4)),(['Random Forest', 'SVM', 'XGBoost', 'Ensemble']))


plt.plot(theta, acc_random, label = 'Accuracy')
plt.plot(theta, bal_acc_random, label = 'Balanced Accuracy')
plt.plot(theta, roc_auc_random, label = 'ROC AUC')
plt.plot(theta, precision_random, label = 'Precision')
plt.plot(theta, recall_random, label = 'Recall')

plt.legend(loc=3)
             
plt.title("Performance Metrics on Random Splitter")
```

```python
# Please write your code below:
metrics = ['Accuracy', 'Balanced accuracy','ROC_AUC', 'Precision', 'Recall']
plt.figure(figsize=(20,10))
plt.subplot(polar=True)
theta = np.linspace(0,2*np.pi, 4)
lines, labels = plt.thetagrids(range(0, 360, int(360/4)),(['Random Forest', 'SVM', 'XGBoost', 'Ensemble']))


plt.plot(theta, acc_scaffold, label = 'Accuracy')
plt.plot(theta, bal_acc_scaffold, label = 'Balanced Accuracy')
plt.plot(theta, roc_auc_scaffold, label = 'ROC AUC')
plt.plot(theta, precision_scaffold, label = 'Precision')
plt.plot(theta, recall_scaffold, label = 'Recall')

plt.legend(loc=3)
             
plt.title("Performance Metrics on Scaffold Splitter")
```

```python
N = 5
ind = np.arange(N) 
width = 0.2
plt.figure(figsize=(20,10))

RF = [acc_random[0], bal_acc_random[0], roc_auc_random[0],precision_random[0],recall_random[0]]
bar1 = plt.bar(ind, RF, width, color = 'r')
  
SVM = [acc_random[1], bal_acc_random[1], roc_auc_random[1] ,precision_random[1],recall_random[1]]
bar2 = plt.bar(ind+width, SVM, width, color='g')
  
XGBoost = [acc_random[2], bal_acc_random[2], roc_auc_random[2],precision_random[2],recall_random[2]]
bar3 = plt.bar(ind+width*2, XGBoost, width, color = 'b')

ensemble = [acc_random[3],bal_acc_random[3],roc_auc_random[3],precision_random[3],recall_random[3]]
bar4 = plt.bar(ind+width*3, ensemble, width, color = 'y')

  
plt.xlabel("Metric")
plt.ylabel('Score')
plt.ylim([0.8, 1.01])

plt.title("Performance Metrics on Random Splitter")  
plt.xticks(ind+width,['Accuracy', 'Balanced Accuracy', 'ROC AUC', 'Precision', 'Recall'])
plt.legend( (bar1, bar2, bar3, bar4), ('Random Forest', 'SVM', 'XGBoost', 'Ensemble') )
plt.show()
```
