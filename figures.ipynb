```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Accuracies in the order (Random Forest, SVM, XGBoost, Ensemble)
acc_random = [0.9538152610441767, 0.963855421686747, 0.9558232931726908,0.9618473895582329]
acc_scaffold = [0.8012048192771084,0.8644578313253012, 0.8674698795180723,0.8614457831325302]
```

```python
#Balanced Accuracies in the order (Random Forest, SVM, XGBoost, Ensemble)
bal_acc_random = [0.8826530612244898,0.9158673469387755,0.9070153061224491,0.910765306122449]
bal_acc_scaffold = [0.777027027027027,0.8499559341950647,0.8520123384253819,0.8445945945945945]
```

```python
#ROC AUC in the order (Random Forest, SVM, XGBoost, Ensemble)
roc_auc_random = [0.8826530612244897, 0.9158673469387756, 0.9070153061224491, 0.910765306122449 ]
roc_auc_scaffold = [0.777027027027027,0.8499559341950647, 0.8520123384253818,0.8445945945945946]
```

```python
#Precision in the order (Random Forest, SVM, XGBoost, Ensemble)
precision_random = [0.9456264775413712,0.961352657004831, 0.9587378640776699,0.9590361445783132]
precision_scaffold = [0.736,0.8116591928251121, 0.8097345132743363, 0.8]
```

```python
#Recall in the order (Random Forest, SVM, XGBoost, Ensemble)
recall_random = [1.0, 0.995, 0.9875, 0.995]
recall_scaffold = [1.0, 0.9836956521739131, 0.9945652173913043, 1.0]
```

```python
# Please write your code below:
metrics = ['Accuracy', 'Balanced accuracy','ROC_AUC', 'Precision', 'Recall']
plt.figure(figsize=(20,10))
plt.subplot(polar=True)
theta = np.linspace(0,2*np.pi, 4)
lines, labels = plt.thetagrids(range(0, 360, int(360/4)),(['Random Forest', 'SVM', 'XGBoost', 'Ensemble']))


plt.plot(theta, acc_random, label = 'Accuracy')
plt.plot(theta, bal_acc_random, label = 'Balanced Accuracy')
plt.plot(theta, roc_auc_random, label = 'ROC AUC')
plt.plot(theta, precision_random, label = 'Precision')
plt.plot(theta, recall_random, label = 'Recall')

plt.legend(loc=3)
             
plt.title("Performance Metrics on Random Splitter")
```

```python
# Please write your code below:
metrics = ['Accuracy', 'Balanced accuracy','ROC_AUC', 'Precision', 'Recall']
plt.figure(figsize=(20,10))
plt.subplot(polar=True)
theta = np.linspace(0,2*np.pi, 4)
lines, labels = plt.thetagrids(range(0, 360, int(360/4)),(['Random Forest', 'SVM', 'XGBoost', 'Ensemble']))


plt.plot(theta, acc_scaffold, label = 'Accuracy')
plt.plot(theta, bal_acc_scaffold, label = 'Balanced Accuracy')
plt.plot(theta, roc_auc_scaffold, label = 'ROC AUC')
plt.plot(theta, precision_scaffold, label = 'Precision')
plt.plot(theta, recall_scaffold, label = 'Recall')

plt.legend(loc=3)
             
plt.title("Performance Metrics on Scaffold Splitter")
```

```python
N = 3
ind = np.arange(N) 
width = 0.2
plt.figure(figsize=(20,10))

RF = [acc_random[0], bal_acc_random[0], roc_auc_random[0]]
bar1 = plt.bar(ind, RF, width, color = 'r')
  
SVM = [acc_random[1], bal_acc_random[1], roc_auc_random[1]]
bar2 = plt.bar(ind+width, SVM, width, color='g')
  
XGBoost = [acc_random[2], bal_acc_random[2], roc_auc_random[2]]
bar3 = plt.bar(ind+width*2, XGBoost, width, color = 'b')

ensemble = [acc_random[3],bal_acc_random[3],roc_auc_random[3]]
bar4 = plt.bar(ind+width*3, ensemble, width, color = 'y')

  
plt.xlabel("Metric")
plt.ylabel('Score')
plt.ylim([0.8, 1.01])

plt.title("Performance Metrics on Random Splitter")  
plt.xticks(ind+width,['Accuracy', 'Balanced Accuracy', 'ROC AUC'])
plt.legend( (bar1, bar2, bar3, bar4), ('Random Forest', 'SVM', 'XGBoost', 'Ensemble') )
plt.show()
```
